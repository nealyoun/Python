# replace vs. str.replace

```python
import seaborn as sns
import pandas as pd
mpg = sns.load_dataset("mpg")
```

```python
## 완전히 값이 같아야지만, 일치해야지만 변경가능
mpg.replace("torino", "______")

```

```python
## Series 에만 적용가능
## 일부만 같아도 변경 가능
mpg["name"].str.replace("torino","_______")
```
